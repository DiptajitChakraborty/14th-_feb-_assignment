```python
#Q2
#Ans:- threading module is used for creating, controlling and managing threads in python. This enables efficient utilization of the resources as the threads share the data space and memory.It also causes reduction of time consumption.

# active_count() , current_thread() and enumerate() method from Threading Module
import threading

def thread1(i):
    print(" Number of active threads:", threading.active_count())
thread1 = threading.Thread(target=thread1, args=(10,))

thread1.start()

#2
def thread_1(i):
    print("Active current thread right now:", (threading.current_thread()))
thread2 = threading.Thread(target=thread_1, args=(1,))



thread2.start()

#3
def thread_1(i):
    print("Threads alive when thread_1 executes:")
    print(*threading.enumerate(), sep = "\n")
        
    print()

thread1 = threading.Thread(target=thread_1, args=(10,))


print("Threads alive in the starting:", threading.enumerate())
print()

thread1.start()

```

```python
#Q3
#Ans:-

#1.
#The run() method executes any target function belonging to a given thread object that is now active. It normally executes in the background after the start() method .
import threading 
def test(id):
    print("program start %d"% id)
    
t = [threading.Thread(target=print, args=(i,)) for i in range(5)]
for e in t :
    e.run()
    
#2
#start() method is an inbuilt method of the Thread class of the threading module in Python. It is used to start a thread's activity.
def test(id):
    print("program start %d"% id)
    
t = [threading.Thread(target=print, args=(i,)) for i in range(5)]
for e in t :
    e.start()
 

#3
#the calling thread gets blocked until the thread object (on which the thread is called) gets terminated. The thread objects can terminate under any one of the following conditions:

#Either normally.
#Through an ill-handled exception.
#Till the optional timeout occurs.
#Hence the join() method indicates wait till the thread terminates
# SuperFastPython.com
# example of joining a thread
import time 
def task():
    time.sleep(1)
    print('first text')
 
thread = threading.Thread(target=task)
thread.start()
print("Waiting for thread to terminate")
thread.join()
 
    
#4
#is_alive() method is an inbuilt method of the Thread class of the threading module in Python. It uses a Thread object, and checks whether that thread is alive or not, ie, it is still running or not. This method returns True before the run() starts until just after the run() method is executed.

import time
import threading

def thread_1(i):
    time.sleep(1)
    print('Value by Thread 1:', i)
thread1 = threading.Thread(target=thread_1, args=(1,))
print("Is thread1 alive:", thread1.is_alive())
print()

thread1.start()
print("Is thread1 alive:", thread1.is_alive())
```

```python
#Q4
#Ans:- 
import threading
 
 
def print_cube(num):
    print("Cube: {}" .format(num * num * num))
 
 
def print_square(num):
    print("Square: {}" .format(num * num))
 
 
if __name__ =="__main__":
    t1 = [threading.Thread(target=print_square, args=(i,))for i in range(10)]
    t2 = [threading.Thread(target=print_cube, args=(j,))for j in range(10)]
 
    for t in t1 :
        t.start()
    for n in t2 :
        n.start()
 
 

 
```
